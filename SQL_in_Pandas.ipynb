```python
# imports
import pandas as pd
from sqlalchemy import create_engine

```

```python
# montar datos
from google.colab import drive
drive.mount('/content/drive')

```

```python
# cargar datos: ¡cambia la ruta para que coincida con tu unidad!
school = pd.read_csv('/content/drive/MyDrive/Coding Dojo/Semana 12/School_Details.csv')
# Asegúrate de que estén cargados correctamente
school.head(2)
```

```python
# crear motor sqlite
engine = create_engine('sqlite:///school.db', echo=True)
# crear conexión con el motor
conn = engine.connect()
```

```python
# agregar marco de datos como una tabla en sqlite
sqlite_table = "school_details"
school.to_sql(sqlite_table, conn, if_exists='fail')
```

```python
# Escribe tu consulta SQL en esta cadena
sql = """
SELECT EstablishmentName, SchoolLevel, Website
FROM school_details 
"""
# Lee la consulta sql en pandas para ver los resultados
df = pd.read_sql_query(sql, engine)
df.head()
```

```python
# Escribe tu consulta SQL en esta cadena
sql = """
SELECT EstablishmentName, [DistrictHQDistance (in Km)] 
FROM school_details
WHERE [DistrictHQDistance (in Km)] > 100
"""
# Lee la consulta sql en pandas para ver los resultados
df = pd.read_sql_query(sql, engine)
df.head()
```

```python
# Escribe tu consulta SQL en esta cadena
sql = """
SELECT EstablishmentName, SchoolLevel, [DistrictHQDistance (in Km)] 
FROM school_details
WHERE SchoolLevel = "JHS"
"""
# Lee la consulta sql en pandas para ver los resultados
df = pd.read_sql_query(sql, engine)
df.head()
```

```python
# Escribe tu consulta SQL en esta cadena
sql = """
SELECT EstablishmentName, SchoolLevel, [DistrictHQDistance (in Km)] 
FROM school_details
WHERE EstablishmentName LIKE "%Junior High%"
"""
# Lee la consulta sql en pandas para ver los resultados
df = pd.read_sql_query(sql, engine)
df.head()
```

```python
#Muestra NombreEstablecimiento, NivelEscolar, DistanciaDistrito (en km) de colegios cuyo nombre comience con la letra “C”. 
#Ordena los resultados de forma alfabética por nombre de colegio  y limita los resultados de la consulta a 10 líneas.

# Escribe tu consulta SQL en esta cadena
sql = """
SELECT EstablishmentName, SchoolLevel, [DistrictHQDistance (in Km)] 
FROM school_details
WHERE EstablishmentName LIKE "C%"
ORDER BY EstablishmentName
LIMIT 10
"""
# Lee la consulta sql en pandas para ver los resultados
df = pd.read_sql_query(sql, engine)
df.head(10)
```

```python
#Muestra los nombres de los colegios de áreas urbanas. Ordena los resultados en orden alfabético inverso.
# Escribe tu consulta SQL en esta cadena
sql = """
SELECT EstablishmentName, LocatedInRuralOrUrban
FROM school_details
WHERE LocatedInRuralOrUrban ="Urban Area"

"""
# Lee la consulta sql en pandas para ver los resultados
df = pd.read_sql_query(sql, engine)
df.head()
```
